# 2. API 문서의 소스 찾아 쓰는 능력 체크 문제
> Question: 아래의 list 안의 12가지의 과일/채소 중 3종류 씩 4가지 경우로 나누어 스무디를 만들려고한다. <br> 
vegi = ['apple', 'avocado', 'blackberry', 'pineapple', 'carrot', 'banana', 'tomato', 'pumpkin', 'strawberry', 'beetroot', 'celery', 'blueberry'] <br> <br>
단, 3개의 조합 중 한번 사용한 과일/채소는 한번 이상 사용이 불가능 하다. <br>
(무슨뜻이냐하면? 예를들어 ('apple', 'apple', 'blackberry') 안됨) <br>
출력값으로 각 조합에 번호를 붙여 전부 다 출력하세요. <br><br>
출력값 예시: <br>
1 번: ('apple', 'avocado', 'blackberry')<br>
2 번: ('apple', 'avocado', 'pineapple')<br>
3 번: ('apple', 'avocado', 'carrot')<br>
4 번: ('apple', 'avocado', 'banana')<br>
5 번: ('apple', 'avocado', 'tomato')<br>
등등... 12가지를 사용한 모든 조합 <br> <br>
*힌트1: <br>
url: https://docs.python.org/3.4/library/itertools.html#itertools <br>
위의 url 주소를 참조하여 다음 문제에 맞는 function을 찾아 사용하세요. <br>
*힌트2: 총 가능한 경우의 수 12C3= 12!/3!(12 - 3)!




```python
vegi = ['apple', 'avocado', 'blackberry', 'pineapple', 'carrot', 'banana', 'tomato', 'pumpkin', 'strawberry', 'beetroot', 'celery', 'blueberry'] 
for i in range(12):
    for j in range(11):
        for k in range(10):
            vegiv=vegi[i],vegi[j+1],vegi[k+2]
            
            print(vegiv)
```

# 3. 행렬의 곱
>아래의 행렬 X 와 행렬 Y의 곱을 도출하는 문을 만드세요. <br>
원하시는 분은 comprehension을 사용하셔도 좋지만 아직은 어려우니 그냥 푸시는걸 추천 드려요.

```python
#2*3 행렬
X = [[1, 2, 3],
     [4 ,5, 6]]

#3*2 행렬
Y = [[1, 2],
     [3 ,4],
     [5 ,6]]
#2*2 행렬
result = [[0,0],
         [0,0]]
```

```python
for a in range(3):
    for b in range(2):
        result[0][b]+=X[0][a]*Y[a][b]
        result[1][b]+=X[1][a]*Y[a][b]
print(result)
```

# 4. reduce() 를 활용한 문제
>아래의 reduce(lambda x, y: x+y, [1, 2, 3, 4]) <br>
의 결과값은 10입니다. <br>
결과값은 ((1+2)+3)+4) 와 같은 방식으로 도출되었습니다. <br>
이와 같은 방식으로 x-y, x*y, x/y 를 활용한 예시를 만들고 답은 어떻게 도출 되었는지 위처럼 설명하세요.<br>

```python
from functools import reduce
reduce(lambda x, y: x+y, [1, 2, 3, 4])
```

```python
# ((10-5)-2)-1)
from functools import reduce
reduce(lambda x, y: x-y, [10, 5, 2, 1])
```

```python
# ((2*2)*2)*2)
from functools import reduce
reduce(lambda x, y: x*y, [2, 2, 2, 2])
```

### 문제 1

아래 주석의 설명을 보고, '매우' 공정한 주사위를 만들어 보세요<br>
만들어진 주사위가 공정한 주사위인지 여부 확인을 위해, 10000번 주사위를 굴려 각 숫자가 나온 빈도를 count해 주세요

```python
'''
    함수 또는 메소드는 아래의 조건을 만족해야 합니다.
    
    
    1. input 값으로 0 ~ 6 까지의 정수와 0 ~ 100까지의 정수를 받습니다.
        첫 번째 정수는 동작 모드를 나타냅니다.
        두 번째 정수는 확률을 조작합니다.
        
        예) f = unBiasedDice()
            while True:
                f.main(동작모드, 확률)
        
    
    2. 첫 번째 정수가 0 인 경우:
        두 번째 정수의 값에 관계없이
        1 ~ 6 까지 숫자가 출력 될 확률이 각각 1/6 이어야 합니다.
        
        예) f = unBiasedDice()
            while True:
                f.main(0, 99)
                -> 1 ~ 6 까지의 숫자가 균등하게 출력되어야 합니다.
    
    
    3. 첫 번째 정수가 0 이 아닌 경우:
        첫 번째 정수의 값이 출력 될 학률이
        두 번째 정수의 값으로 반영되어야 합니다.
        나머지 숫자가 나올 확률은 균등한 주사위가 됩니다.
        
        예) f = unBiasedDice()
            while True:
                f.main(6, 95)
                -> 1 은 1% 의 확률로 출력되어야 한다.
                -> 2 은 1% 의 확률로 출력되어야 한다.
                -> 3 은 1% 의 확률로 출력되어야 한다.
                -> 4 은 1% 의 확률로 출력되어야 한다.
                -> 5 은 1% 의 확률로 출력되어야 한다.
                -> 6 은 95% 의 확률로 출력되어야 한다.

'''

from random import randint
import time

class unBiasedDice:
    def __init__(self):
        self.output_val = 0
        self.input_val = 0
        self.justice = 0

    def main(self, input_val, percent):
        self.input_val = input_val
        self.percent = percent
        self.a = [1,2,3,4,5,6]
        self.a.remove(input_val)
        self.b = []
        
        for i in range(1,10):
            self.a.extend(self.a)
            
        if (percent < 100 and percent >= 0):
            self.justice = int((len(self.a) * self.percent) / (100 - self.percent))        
        elif(percent >= 100):
            self.justice = 1
            self.a = [input_val]
        
        if (self.input_val == 0):
            self.output_val = randint(1,6)
            print(self.output_val)
        else:
            for i in range(1,self.justice):
                self.b.append(self.input_val)
            self.a.extend(self.b)
            print(self.a[randint(0,len(self.a)-1)])
            time.sleep(0.1)
                    
if __name__ == "__main__":
    f = unBiasedDice()
    while True:
        f.main(3, 100)
```

```python
#답안
def print_menu():
    print('1. 전화 번호 출력')
    print('2. 전화 번호 추가')
    print('3. 전화 번호 삭제')
    print('4. 전화 번호 찾기')
    print('5. 종료')

def print_dic():
    if not numbers:
        print("전화번호부가 비어 있습니다\n")
    else:
        print("등록된 전화번호 목록입니다\n")
        print("이름 ( 전화번호 )")
        print("="*20)
        for index in numbers.keys():
            print(index + " ( "+numbers[index]+" )")
        print()

def add_number():
    print("목록에 추가합니다")
    while 1:
        name = input("이름을 입력하세요:")
        if name: break
    while 1:
        phone = input("전화번호를 입력하세요(000-0000-0000):")
        if phone: break

    numbers[name] = phone

    print("목록에 추가되었습니다.")

def remove_number():
    print_dic()

    delete = input("삭제할 이름을 입력하세요:")
    try:
        del numbers[delete]
    except:
        print("잘못된 이름입니다")

def lookup_number():
    look = input("찾을 이름을 입력하세요:")
    try:
        print("찾으시는 이름의 전화번호는 "+ numbers[look] +"입니다.")
    except:
        print("찾는 이름이 없습니다")

if __name__== "__main__":
    numbers = {}
    
    with open("number.txt") as f:
        for line in f.readlines():
            data = line.strip("\n").split(":")
            numbers[data[0]] = data[1]
    
    while 1:
        menu_choice = 0
        print_menu()
        menu_choice = input("처리할 메뉴의 숫자를 입력하세요:")

        if menu_choice == "1":
            print_dic()
        elif menu_choice == "2":
            add_number()
        elif menu_choice == "3":
            remove_number()
        elif menu_choice == "4":
            lookup_number()
        elif menu_choice == "5":
            print("프로그램을 종료합니다")
            
            with open("number.txt", "w") as f:
                for key in numbers.keys():
                    f.write(key+":"+numbers[key]+'\n')
            break
            
        else:
            print("1~5 사이의 올바른 숫자를 입력하세요")
```
